Na máquina local, recompilando os exemplos usando o container. O dir $HOME do container está mapeado para o $HOME da máquina local. Uma cópia do repo TorchFort está em $HOME/torchfort.

O TorchFort é feito para ser recompilado junto com os exemplos toda vez que altera alguma coisa nos exemplos, não existe uma biblioteca separada.

```python
%%bash
export FC=nvfortran
export HDF5_ROOT=/opt/hdf5
export CUDA_PATH=/opt/nvidia/hpc_sdk/Linux_x86_64/24.1/cuda
cd /home/x/torchfort
rm -rf build && mkdir build && cd build
cmake -DCMAKE_INSTALL_PREFIX=/home/x/torchfort/build/install \
      -DTORCHFORT_YAML_CPP_ROOT=/opt/yaml-cpp \
      -DTORCHFORT_BUILD_EXAMPLES=1 \
      -DCMAKE_PREFIX_PATH="`python -c 'import torch;print(torch.utils.cmake_prefix_path)'`;${CUDA_PATH}" \
    ..
make -j
```
