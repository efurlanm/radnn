Optional:

```python
! singularity instance start --nv \
    --writable-tmpfs \
    --no-home \
    --bind $HOME:/mnt/x \
    $HOME/containers/torchfort_v4.sif t01
```

FLAGS

- The --writable-tmpfs flag allocates a filesystem overlay in RAM, which allows writing to the container instance. Everything is lost when the container exits.
- The --no-home flag do NOT mount users home directory if home is not the current working directory.
